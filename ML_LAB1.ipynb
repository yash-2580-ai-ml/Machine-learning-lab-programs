```python
import numpy as np

# Question 1
arr = np.arange(1, 21)
print("Array:", arr)
print("Shape:", arr.shape)
print("Max:", arr.max(), "Min:", arr.min())
print("Mean:", arr.mean(), "Std Dev:", arr.std())
matrix = arr.reshape(4, 5)
print("Matrix:\n", matrix)
print("Second row:", matrix[1])
print("Third column:", matrix[:, 2])
```

```python
# Question 2
A = np.array([10, 20, 30, 40, 50])
B = np.array([5, 10, 15, 20, 25])
print("Addition:", A + B)
print("Subtraction:", A - B)
print("Multiplication:", A * B)
print("Division:", A / B)
print("Sum of A:", A.sum())
```

```python
import pandas as pd

# Question 3
data = {
    'Roll No': [101, 102, 103, 104, 105],
    'Name': ['Ravi', 'Priya', 'Kiran', 'Anu', 'Sita'],
    'Marks': [85, 92, 78, 88, 95]
}
df = pd.DataFrame(data)
print(df)
print("First 3 records:\n", df.head(3))
print("Average Marks:", df['Marks'].mean())
print("Highest Marks student:", df.loc[df['Marks'].idxmax(), 'Name'])
print("Lowest Marks student:", df.loc[df['Marks'].idxmin(), 'Name'])

def get_grade(marks):
    if marks >= 90: return 'A'
    elif marks >= 80: return 'B'
    else: return 'C'

df['Grade'] = df['Marks'].apply(get_grade)
print(df)
```

```python
# Question 4
products = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Price': [50000, 500, 1200, 10000],
    'Quantity': [10, 50, 30, 15]
}
df_prod = pd.DataFrame(products)
df_prod['Total Value'] = df_prod['Price'] * df_prod['Quantity']
print(df_prod)
highest = df_prod.loc[df_prod['Total Value'].idxmax(), 'Product']
print("Highest total value product:", highest)
print("Overall inventory value:", df_prod['Total Value'].sum())
```

```python
from sklearn.datasets import load_iris
import pandas as pd

# Question 5
iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

print("First 5 records:\n", df_iris.head())
print("Shape:", df_iris.shape)
print("Columns:", df_iris.columns.tolist())
print("Missing values:\n", df_iris.isnull().sum())
print("Summary statistics:\n", df_iris.describe())
print("Samples per class:\n", df_iris['target'].value_counts())
```
